```python
from bs4 import  BeautifulSoup
```

```python
import requests # works at the background to request data
```

```python
# find the trending python certification courses
url  = requests.get('https://www.codespaces.com/best-python-certifications-courses-tutorials.html')
```

```python
url
```

```python
htmltext = url.text
```

```python
print(htmltext)
```

```python
file = open('/content/drive/MyDrive/Data Mining & Machine Learning/pyhtoncourselist.html','w') # creates a file and opens it in  write mode
```

```python
# write contents into file
file.write(htmltext)
```

```python
# next day you want to read the data for processing
with open('/content/drive/MyDrive/Data Mining & Machine Learning/pyhtoncourselist.html', 'r') as html_file:
  content = html_file.read()
```

```python
print(content)
```

```python
# prettify() store the tags into objects

soup = BeautifulSoup(content, 'lxml')
```

```python
print(soup.prettify())
```

```python
# locate the course details from header h2
head_2 =soup.find_all('h2')
print(head_2)
```

```python
# extract all the data course names 
for course in head_2:
  #if course.text[0].isdigit:
  print(course.text)
  
```
